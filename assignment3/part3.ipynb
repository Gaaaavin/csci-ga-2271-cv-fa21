```python
import numpy as np
import scipy.io as io
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
sfm_points = io.loadmat("sfm_points.mat")["image_points"]

centroids = np.mean(sfm_points, axis=1, dtype=np.float64)
centered_points = sfm_points - np.expand_dims(centroids, axis=1)
W = np.transpose(centered_points, (2, 0, 1)).reshape(20, 600)

u, s, v = np.linalg.svd(W)
M = u.T[:, :3] @ np.diag(s[:3])
point_real = v.T[:, :3]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(point_real[:, 0], point_real[:, 1], point_real[:, 2])
```
